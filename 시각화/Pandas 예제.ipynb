### oo중학교 1학년 1반의 중간고사 평균 성적 (국어, 영어, 수학, 과학, 사회) 점수를 Series로 생성
 - (국어 : 75점   영어 : 80점   수학 : 90점   과학 : 95점   사회  70점 )


```python
score = pd.Series([75,80,90,95,70], 
                  index = ['국어','영어','수학','과학','사회'])
score
```

```python
score.name = '중간고사성적'
score.index.name = '과목'
```

```python
print(score[0:3])
print(score["국어":"수학"])
```

```python
# 중간고사 평균 성적이 85점 이상인 과목만 뽑기
score[score >= 85]
```

```python
# 중간고사 평균 성적이 80점 이상 90점 이하인 과목만 뽑기
score[(score >= 80)  & (score <= 90)]
```

```python
#5개 과목의 평균점수, 합계점수, 최대값을 출력
import numpy as np
print(np.mean(score))
print(np.sum(score))
print(np.max(score))
```

## pivot table

```python
#연도별 메달 수는? -> 결측치의 값을 0으로 채우시오.
a=kr.pivot_table(index='Year',columns='Medal',aggfunc='size').fillna(0)
a
```

# 5세이하 어린이 사망률

```python
import pandas as pd
df= pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/mergeTEst.csv')
df
```

```python
df1=df.iloc[:4,:]
df2=df.iloc[4:,:]

display(df1)
display(df2)
```

```python
#df1 과 df2를 하나의 데이터프레임으로 만들기
t = pd.concat([df1,df2])
t
```

```python
df3=df.iloc[:2,:4]
df4=df.iloc[5:,3:]

display(df3)
display(df4)
```

```python
#df3과 df4를 합치기, 둘다 포함되는 연도에 대해서만 고려하기(공동칼럼)
s= pd.concat([df3,df4], join = 'inner')    
s                                                
```

```python
#df3과 df4를 합치기, 단, 모든 칼럼 포함 결측치는 0으로 대체
v = pd.concat([df3,df4]).fillna(0)         
v
```
