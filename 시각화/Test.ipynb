```python
data= pd.read_csv('Traffic_Accident_2017.csv',encoding='cp949')
data.head(3)
```

```python
fig, (ax1) = plt.subplots(ncols=1, figsize=(8,5))

sns.scatterplot(data=data, x='경도', y='위도',ax=ax1, hue='주야',alpha = 0.1)
plt.xlim(123,133) 
plt.xlabel('LON')
plt.ylabel('LAT')
plt.legend()
plt.show()
```

```python
Xn=data[data['주야']=='야간']['경도']
Yn=data[data['주야']=='야간']['위도']
Xd=data[data['주야']=='야간']['경도']
Yd=data[data['주야']=='야간']['위도']

plt.scatter(Xn,Yn, c='blue',alpha=0.1,label='Night')
plt.scatter(Xd,Yd, c='red',alpha=0.1,label='Day')

plt.xlim(123,133) 
plt.xlabel('LON')
plt.ylabel('LAT')
plt.legend()
plt.show()
```
