# 카테고리 만들기

```python
ages = [0,2,10,21,23,37,31,61,20,41,32,100]
bins = [0,15,25,35,60,99]
labels = ["미성년자","청년",'중년',"장년","노년"]
cats = pd.cut(ages,bins,labels=labels)
```

```python
cats
```

```python
type(cats)
```

```python
cats.categories
```

### 1. age 를 DataFrame으로 만들기

```python
ageArr = pd.DataFrame(ages, columns = ['ages'])
ageArr
```

### 2. 각 연령에 맞게 카테고리를 추가

```python
ageArr['age_cat'] = cats
ageArr
```

# DataFrame 병합

## concat 함수

```python
df1 = pd.DataFrame({'A' : ['A0', 'A1', 'A2', 'A3'],
                    'B' : ['B0', 'B1', 'B2', 'B3'],
                    'C' : ['C0', 'C1', 'C2', 'C3'],
                    'D' : ['D0', 'D1', 'D2', 'D3']},
                   index = [0,1,2,3])

df2 = pd.DataFrame({'A':['A4','A5','A6','A7'],
                   'B':['B4','B5','B6','B7'],
                   'C':['C4','C5','C6','C7'],
                    'D' : ['D4', 'D5', 'D6', 'D7']},
                   index=[4,5,6,7])

df3 = pd.DataFrame({"A" : ["A8", "A9", "A10", "A11"],
                    "B" : ["B8", "B9", "B10", "B11"],
                    "C" : ["C8", "C9", "C10", "C11"],
                    "D" : ["D8", "D9", "D10", "D11"]},
                   index = [8,9,10,11])
```

```python
result = pd.concat([df1,df2,df3])
result
```

```python
result = pd.concat([df1,df2,df3], keys = ['x','y','z'])       # keys 속성 부여 시 다중 index가 된다.
result
```

```python
result.index
```

```python
result.index.get_level_values(0)          #다중 index확인 (index 값 확인)
```

```python
df4 = pd.DataFrame({"B" : ["B2", "B3", "B6", "B7"],
                    "D" : ["D2", "D3", "D6", "D7"],
                    "F" : ["F2", "F3", "F6", "F7"]},
                   index = [2,3,6,7])
df4
```

#### 행 방향으로 병합하기

```python
result = pd.concat([df1,df4], axis = 1)          # outer join
result
```

```python
result = pd.concat([df1,df4], axis = 0)
result
```

```python
result = pd.concat([df1,df4], axis = 1, join = 'inner')     # 행이 겹치는 부분(0123)
result                                                      #                ( 2367)
```

```python
result = pd.concat([df1,df4], axis = 0, join = 'inner')      # 행이 겹치는 부분(ABCD)
result                                                       #                 (BDF)
```

### concat 함수의 ignore_index 속성
- 기존 인덱스를 무시하고 새로운 인덱스를 부여하는 기능

```python
result = pd.concat([df1,df4],ignore_index = True)  
result                                      
```

```python
result = pd.concat([df1,df4], axis = 0)      
result                                      
```

### 구분칼럼의 값이 발생건수인 데이터만 가져오기

```python
df2015_crime =df2015[df2015['구분'] == '발생건수']

df2015_crime['2015총계'] = df2015_crime.sum(axis = 1)
df2015_crime
```

```python
df2016_crime =df2016[df2016['구분'] == '발생건수']

df2016_crime['2016총계'] = df2016_crime.sum(axis = 1)
df2016_crime
```

```python
df2017_crime =df2017[df2017['구분'] == '발생건수']

df2017_crime['2017총계'] = df2017_crime.sum(axis = 1)
df2017_crime
```

### 데이터 총합 열 생성

```python
s1 = df2015_crime['2015총계']
s3 = df2016_crime['2016총계']
s5 = df2017_crime['2017총계']
```

###  1.전년대비 증감율 계산하기
###  2.이름 변경하기

```python
s2 = (s3 - s1)/s1 * 100
s2.name = '2015-2016 증감율'
# Series의 name값은 DataFrame에서 column값으로 사용됨
```

```python
s4 = (s5 -s3)/s3 * 100
s4.name = '2016-2017 증감율'
```

```python
pd.concat([s1,s2,s3,s4,s5], axis = 1)
```

## Merge

- 공통 칼럼이 존재해야하며, 그 공통 칼럼을 기준으로 데이터프레임들을 병합시킴

```python
employee = [{'empno':1, 'ename':'kim', 'dept':1},
            {'empno':2, 'ename':'lee', 'dept':2},
            {'empno':3, 'ename':'park', 'dept':1},
            {'empno':4, 'ename':'song', 'dept':3},
            {'empno':5, 'ename':'min', 'dept':2}
           ]
dept=[{'dept':1, 'deptname':'관리직'},
      {'dept':2, 'deptname':'영업직'},
      {'dept':3, 'deptname':'개발직'}
     ]
info =[{'empno':1, 'addr':'서울시','phone':'010-1111-1111'},
       {'empno':3, 'addr':'부산시','phone':'010-2222-2222'},
       {'empno':2, 'addr':'광주시','phone':'010-3333-3333'},
       {'empno':5, 'addr':'광주시','phone':'010-4444-4444'},
       {'empno':4, 'addr':'광주시','phone':'010-5555-5555'}
      ]
```

```python
emp = pd.DataFrame(employee)
dept = pd.DataFrame(dept)
info = pd.DataFrame(info)
```

```python
pd.merge(emp,dept)
```

```python
mg = pd.merge(dept,emp,on='dept')            
mg.drop(columns=['dept'], inplace= True)     # inplace = True : 데이터프레임이 실제로 바뀜, 변수를 지정하지 않고도 mg이름을 그대로 사용해서 drop 사용
mg
```
