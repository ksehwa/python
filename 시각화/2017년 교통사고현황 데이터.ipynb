```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('Traffic_Accident_2017.csv',encoding='cp949')
data
```

```python
#요일별 사고 건수를 세봅시다
temp=data['요일'].value_counts()
temp
```

```python
y=temp[['월','화','수','목','금','토','일']]
y
```

```python
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname = 'C:/Windows/Fonts/malgun.ttf').get_name()
rc('font',family=font_name)
```

```python
x=y.index
plt.bar(x,y)
plt.ylim(500,620)                             # 범위지정
plt.xlabel('요일')                            # x축 이름지정
plt.ylabel('사고 건수').set_rotation(0)
plt.title('2017년 전국 교통사망사고')
plt.show()
```

```python
#'사고유형_대분류' 칼럼에서 '차대차'만 뽑아오기
car= data[data['사고유형_대분류'] == '차대차']
car
```

```python
#car 에서 발생지시도를 기준으로 합계구하기
place=car[['사상자수','발생지시도']].groupby('발생지시도').sum()                       #groupby: groupby()기준으로 보는것
place
```

```python
x = place.index
y = place.values
plt.plot(x,y)
plt.xlabel('지역')
plt.ylabel('사상자수').set_rotation(0)
plt.title('차vs차 교통사고 사상자 수')
plt.show
```
