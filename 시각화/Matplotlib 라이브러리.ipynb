```python
import matplotlib.pyplot as plt
import numpy as np
```

## Line Plot 그리기

```python
y = [2,4,6]
plt.plot(y)
plt.show
```

```python
x = [1,2,3]
y = [2,4,6]
plt.plot(x,y)
plt.show
```

## Line style 설정하기

```python
x = [1,2,3]
y = [2,4,6]
plt.plot(x,y,ls = '--')
plt.show
```

```python
x = [1,2,3]
y = [2,4,6]
plt.plot(x,y,ls = ':')
plt.show
```

```python
x = [1,2,3]
y = [2,4,6]
plt.plot(x,y,ls = '-.')
plt.show
```

## Marker 찍기

```python
x = [1,2,3]
y = [2,4,6]
plt.plot(x,y,ls = '--', marker = 'o')
plt.show
```

```python
x = [1,2,3]
y = [2,4,6]
plt.plot(x,y,ls = '--', marker = '+')
plt.show
```

```python
x = [1,2,3]                                    # 굵기 바꾸기
y = [2,4,6]
plt.plot(x,y,ls = '--', marker = 'o',lw = 5)
plt.show
```

```python
x = [1,2,3]                                      
y = [2,4,6]
plt.plot(x,y,ls = '--', marker = 'o',lw = 5, color = 'r', markerfacecolor = 'k')    #선색깔 바꾸기
plt.show
```

```python
x = [1,2,3]                                      
y = [2,4,6]
plt.plot(x,y,ls = '--', marker = 'o',lw = 3, color = 'r', markerfacecolor = 'b',markersize = 7)    
plt.show
```

```python
x = np.arange(8)
y = [5,2,3,6,8,7,5,7]
plt.plot(x,y,ls = '--', marker = 'o',color = 'g',markerfacecolor = 'g',markersize = 7 )
```

```python
x = np.arange(8)
y = [5,2,3,6,8,7,5,7]
plt.plot(x,y,ls = '--', marker = 'D',color = 'r',markerfacecolor = 'b' )
```

```python
x = np.arange(8)
y = [5,2,3,6,8,7,5,7]
plt.plot(x,y,ls = '-.', marker = '<',color = 'm', markerfacecolor = 'k' ,markersize = '10')
```

### X축 Y축 변경

```python
x = np.arange(8)
y = [5,2,3,6,8,7,5,7]
plt.plot(x,y,ls = '-.', marker = '<',color = 'm', markerfacecolor = 'k' ,markersize = '10')
plt.xlim(-2,10)          # 범위 지정
plt.ylim(-5,12)
plt.show()
```

### 그리드 그리기

```python
x = np.arange(8)
y = [5,2,3,6,8,7,5,7]
plt.plot(x,y,ls = '-.', marker = '<',color = 'm', markerfacecolor = 'k' ,markersize = '10')
plt.xlim(-2,10)          
plt.ylim(-5,12) 
plt.grid()              #plt.grid : 격자 그리기
plt.show()
```

```python
x = np.arange(8)
y = [5,2,3,6,8,7,5,7]
plt.plot(x,y,ls = '--', lw = 5, marker = 'o',  ms =15, mec = 'g', markerfacecolor = 'r')
plt.xticks([0,3,6])            # 틱 설정(눈금 설정)
plt.yticks([1,5,9])   
plt.show()
```

## 여러개 plot 그리기

```python
# 여러개 그래프 그리기
# 한 그림안에 여러개 그래흐 그리기 
x = [1,2,3,4]
y = [2,4,6,8]
z = [3,6,9,12]
plt.plot(x)    # 그래프 그리기
plt.plot(y)  
plt.plot(z)
plt.show()
```

## 범례 표시

```python
plt.plot(x, label = 'x')
plt.plot(y, label = 'y')
plt.plot(z, label = 'z')
plt.legend()
plt.show()
```

```python
plt.plot(x, label = '가')
plt.plot(y, label = '나')
plt.plot(z, label = '다')
plt.legend()
plt.show()
```

## x축 y축 title 표시

```python
plt.plot(x, label = 'x')
plt.plot(y, label = 'y')
plt.plot(z, label = 'z')
plt.xlabel('x축')
plt.xlabel('y축')
plt.legend()
plt.title('범례 표시')
plt.show()
```

```python
plt.plot(x, label = 'x')
plt.plot(y, label = 'y')
plt.plot(z, label = 'z')
plt.xlabel('x축')
plt.xlabel('y축').set_rotation(50)
plt.title('범례 표시')
plt.legend()
plt.show()
```
