```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import pandas as pd
import numpy as np
import mglearn
import matplotlib.pyplot as plt
plt.rcParams['font.family']='Malgun Gothic'
plt.rcParams['axes.unicode_minus']=False
```

```python
wine_df = pd.read_csv('wine_data.csv')
wine_df.head()
```

```python
plt.scatter(wine_df.values[:,0],wine_df.values[:,1])
plt.title('wine data')
plt.xlabel('OD_read')
plt.ylabel('Proline')
plt.show()
```

### K-Means

```python
km = KMeans(3)
km_cluster = km.fit_predict(wine_df)
```

```python
plt.scatter(wine_df.values[:,0],wine_df.values[:,1], c=km_cluster)
plt.title('wine - KMeans')
plt.xlabel('OD_read')
plt.ylabel('Proline')
plt.show()
```

### 계층적 군집분석

```python
a = AgglomerativeClustering(n_clusters=3,linkage='ward')
a_cluster = a.fit_predict(wine_df)
```

```python
plt.scatter(wine_df.values[:,0],wine_df.values[:,1],c=a_cluster)
plt.title('wine-Agglomerative')
plt.xlabel('OD_read')
plt.ylabel('Proline')
plt.show()
```

#### 실루엣 지수
- 얼마나 클러스터에 데이터가 잘 맞는지 분석해주는 지표
- 값은 -1에서 1까지 숫자로 나타남
- 값이 1에 가까울수록 완벽하게 클러스터가 구분이 됐다는 뜻

```python
print('실루엣지수 : \n')
print('K-Means : ' , silhouette_score(wine_df,km_cluster))
print('계층적군집분석_agg : ', silhouette_score(wine_df,a_cluster))
```
